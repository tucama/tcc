```python
import pandas as pd
import numpy as np
import os
import re
```

### Get all .lrd files and group tem per pdb

```python
from collections import defaultdict

files = os.listdir(".")

files_per_pdb = defaultdict(list)
regex_pdb = re.compile("([1-9a-z]{4})_")

for file in files:
    if re.match(regex_pdb, file):
        pdb = file[:4]
        files_per_pdb[pdb].append(file.rsplit('.', 1)[0])
```
